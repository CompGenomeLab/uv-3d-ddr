```python
import pandas as pd
import ptitprince as pt

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/cta/users/vkaya/hi-c/work/hela/notebooks/ab/compartments_top20.tsv", sep="\t")
```

```python
fig, ax = plt.subplots(ncols=3, figsize=(20, 5), sharey=False, sharex=True)
ss_log2 = "$log_{2}$"
s_colors = ['#465775', '#A63446', '#F5B841', '#9DBBAE']
comp_mapping = {"aa": "Intra A-A", "bb": "Intra B-B", "ab": "Inter A-B"}

sigma = .2

for i, k in enumerate(["aa", "bb", "ab"]):
    v = pt.RainCloud(x = "sample", y = k, data = df, palette = s_colors, bw = sigma,
                 width_viol = .6, ax = ax[i], orient = "vertical", dodge = True, pointplot = True, move = .2)
    v.set_title(f"{comp_mapping[k]}", fontsize=20)
    if i == 0:
        v.set_ylabel(f"{ss_log2} Observed / Expected", fontsize=18)
    else:
        v.set_ylabel("")

    if i != 2:
        v.set_ylim(0, 0.42)

    v.set_xlabel("")

    v.set_xticklabels(v.get_xticklabels(), fontsize=20)
    v.tick_params(axis='both', which='major', labelsize=14)
    v.tick_params(axis='both', which='minor', labelsize=20)

plt.savefig("compartments_top20.svg",  bbox_inches="tight")
plt.savefig("compartments_top20.png",  bbox_inches="tight", dpi=300)
```

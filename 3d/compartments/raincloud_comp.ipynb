```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import ptitprince as pt
```

```python
clusters = [3,4,6,9,10]

fig, axs = plt.subplots(1, len(clusters), figsize=(32, 5), gridspec_kw={'wspace': 0.5})
s_colors = ['#465775', '#A63446', '#F5B841', '#9DBBAE']
for i, cluster in enumerate(clusters):
    df = pd.read_csv(f'/home/carlos/oldies/manuscripts/review/comp_switch_other_clusters/cluster_{cluster}_gene_counts.tsv', sep='\t')
    ax = axs[i]

    print(f"Cluster {cluster} has {len(df.geneId.unique())} genes")
    pt.RainCloud(x = 'variable', y = 'value', data = df, palette = s_colors, bw = .2, width_viol = .5, orient = "v", pointplot = "black" , alpha = 1, ax = ax, move = .2)
    ax.set_xlabel('')
    if i == 0:
        ax.set_ylabel('Expression Profile', fontsize=18)
    else:
        ax.set_ylabel('')
    ax.tick_params(axis='both', which='major', labelsize=15)
    ax.tick_params(axis='both', which='minor', labelsize=15)

fig.savefig('/home/carlos/oldies/manuscripts/review/comp_switch_other_clusters/cluster_rainclouds.svg', format='svg', bbox_inches = 'tight')
fig.savefig('/home/carlos/oldies/manuscripts/review/comp_switch_other_clusters/cluster_rainclouds.png', dpi = 300, bbox_inches = 'tight')
```

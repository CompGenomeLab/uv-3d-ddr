```python
import pandas as pd
from statannotations.Annotator import Annotator
import ptitprince as pt
import matplotlib.pyplot as plt
```

```python
plot_DF = pd.read_csv("/cta/users/vkaya/hi-c/work/hela/notebooks/boundaries/boundary_strength_raincloud.tsv", sep="\t")
mapping_dict = {"Control": "Control", "12min": "12 min", "30min": "30 min", "60min": "60 min"}
plot_DF["sample1"] = plot_DF["sample1"].map(mapping_dict)
plot_DF["sample1"] = plot_DF["sample1"].astype("category")
plot_DF.sample1 = plot_DF["sample1"].cat.reorder_categories(["Control", "12 min", "30 min", "60 min"])
```

```python
s_colors = ['#465775', '#A63446', '#F5B841', '#9DBBAE']

fig, ax = plt.subplots(figsize=(5, 5))

pt.RainCloud(x = 'sample1', y = 'bs1', data = plot_DF, palette = s_colors, bw = .2, width_viol = .5, orient = "v", pointplot = "black" , alpha = .75, ax = ax)
ax.set_xlabel('Timepoints', fontsize=12)
ax.set_ylabel('Boundary Strength', fontsize=12)

# remove y ticks after 3
ax.set_yticks(ax.get_yticks()[::2])


ax.axhline(y=plot_DF.loc[plot_DF['sample1'] == 'Control', 'bs1'].mean(), color='black', linestyle='--', linewidth=.5)

for i, sample in enumerate(['Control', '12min', '30min', '60min']):
    ax.text(
        i, 
        plot_DF.loc[plot_DF['sample1'] == sample, 'bs1'].max() + 0.08, 
        str(len(plot_DF.loc[plot_DF['sample1'] == sample])), 
        horizontalalignment='center', size='large', color='black', weight='regular')

pairs=[("Control", "12 min"), ("Control", "30 min"), ("Control", "60 min"), ("12 min", "30 min"), ("12 min", "60 min"), ("30 min", "60 min")]

annotator = Annotator(ax, pairs, data=plot_DF, x="sample1", y="bs1", order=['Control', '12 min', '30 min', '60 min'])
annotator.configure(test='t-test_ind', text_format='star', loc='inside')
annotator.apply_and_annotate()
```
